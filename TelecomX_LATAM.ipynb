#ðŸ“Œ ExtracciÃ³n

```python
#importanto la biblioteca de pandas

import pandas as pd
```

```python
#importando el archivo JSON

df = pd.read_json('TelecomX_Data.json')

df
```

```python
#se cuenta con informaciÃ³n anidad se debe normalizar cada columna

df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

#unimos los df normalizdos

df_normalizado = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)

df_normalizado

```

#ðŸ”§ TransformaciÃ³n

```python
#nombres de las columnas y su tipo

df_normalizado.info()
```

```python
#importamos la biblioteca numpy para transformar datos

import numpy as np
```

```python
#transformamos el tipo de datos de algunas columnas

df_normalizado['Churn'] = df_normalizado['Churn'].str.strip().replace('', np.nan)
df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].str.strip().replace('', np.nan).astype(float)

df_normalizado
```

```python
df_normalizado.info()
```

```python
#identificamos nuestra columna llave unica

[col for col in df_normalizado.columns if df_normalizado[col].is_unique]
```

```python
#creamos la columna cuenta diaria

df_normalizado['Cuentas_Diarias'] = df_normalizado['Charges.Monthly'] / 30

df_normalizado
```

#### **Diccionario de datos**

- `customerID`: nÃºmero de identificaciÃ³n Ãºnico de cada cliente
- `Churn`: si el cliente dejÃ³ o no la empresa
- `gender`: gÃ©nero (masculino y femenino)
- `SeniorCitizen`: informaciÃ³n sobre si un cliente tiene o no una edad igual o mayor a 65 aÃ±os
- `Partner`: si el cliente tiene o no una pareja
- `Dependents`: si el cliente tiene o no dependientes
- `tenure`: meses de contrato del cliente
- `PhoneService`: suscripciÃ³n al servicio telefÃ³nico
- `MultipleLines`: suscripciÃ³n a mÃ¡s de una lÃ­nea telefÃ³nica
- `InternetService`: suscripciÃ³n a un proveedor de internet
- `OnlineSecurity`: suscripciÃ³n adicional de seguridad en lÃ­nea
- `OnlineBackup`: suscripciÃ³n adicional de respaldo en lÃ­nea
- `DeviceProtection`: suscripciÃ³n adicional de protecciÃ³n del dispositivo
- `TechSupport`: suscripciÃ³n adicional de soporte tÃ©cnico, menor tiempo de espera
- `StreamingTV`: suscripciÃ³n de televisiÃ³n por cable
- `StreamingMovies`: suscripciÃ³n de streaming de pelÃ­culas
- `Contract`: tipo de contrato
- `PaperlessBilling`: si el cliente prefiere recibir la factura en lÃ­nea
- `PaymentMethod`: forma de pago
- `Charges.Monthly`: total de todos los servicios del cliente por mes
- `Charges.Total`: total gastado por el cliente

```python
#vamos a estimar cuantas suscripciones tiene el cliente

df_normalizado['SuscurpciÃ³n_telefono'] = df_normalizado['PhoneService'].apply(lambda x: 1 if x == 'Yes' else 0)
df_normalizado['SuscurpciÃ³n_internet'] = df_normalizado['InternetService'].apply(lambda x: 0 if x == 'No' else 1)
df_normalizado['SuscurpciÃ³n_seguridad'] = df_normalizado['OnlineSecurity'].apply(lambda x: 1 if x == 'Yes' else 0)
df_normalizado['SuscurpciÃ³n_respaldo'] = df_normalizado['OnlineBackup'].apply(lambda x: 1 if x == 'Yes' else 0)
df_normalizado['SuscurpciÃ³n_proteccion'] = df_normalizado['DeviceProtection'].apply(lambda x: 1 if x == 'Yes' else 0)
df_normalizado['SuscurpciÃ³n_soporte'] = df_normalizado['TechSupport'].apply(lambda x: 1 if x == 'Yes' else 0)
df_normalizado['SuscurpciÃ³n_TV'] = df_normalizado['StreamingTV'].apply(lambda x: 1 if x == 'Yes' else 0)
df_normalizado['SuscurpciÃ³n_peliculas'] = df_normalizado['StreamingMovies'].apply(lambda x: 1 if x == 'Yes' else 0)

df_normalizado
```

```python
#estimamos el total de sucripciones

df_normalizado['Total_suscripciones'] = df_normalizado[['SuscurpciÃ³n_telefono'
                                                       ,'SuscurpciÃ³n_internet'
                                                       ,'SuscurpciÃ³n_seguridad'
                                                       ,'SuscurpciÃ³n_respaldo'
                                                       ,'SuscurpciÃ³n_proteccion'
                                                       ,'SuscurpciÃ³n_soporte'
                                                       ,'SuscurpciÃ³n_TV'
                                                       ,'SuscurpciÃ³n_peliculas'
                                                       ]
                                                      ].sum(axis=1)

df_normalizado
```

```python
df_normalizado.nunique()
```

```python
#validamos la informaciÃ³n de las columnas

df_normalizado['InternetService'].unique()
```

```python
#renombramos algunas columnas

df_normalizado.rename(columns={
    'Churn' : 'Dejo_la_empresa?',
    'gender' : 'Genero',
    'SeniorCitizen' : 'Edad >= 65',
    'Partner' : 'Tiene_Pareja',
    'Dependents' : 'Tiene_Dependientes',
    'tenure' : 'Meses_contrato',
    'Contract' : 'Tipo_contrato',
    'PaperlessBilling' : 'Facturacion_electronica'
}, inplace = True)

df_normalizado
```

```python
# DistribuciÃ³n de abandono por metodo de pago

metodo_pago = (pd.crosstab(df_normalizado['PaymentMethod'], df_normalizado['Dejo_la_empresa?'], normalize = 'index') * 100).round(2)
metodo_pago.plot(kind = 'bar', figsize = (10,6), color = sns.color_palette('pastel'))

plt.title('DistribuciÃ³n de abandono por mÃ©todo de pago', fontsize = 24, fontweight = 'bold')
plt.ylabel('Porcentaje')
plt.xlabel('Metodo de pago')
plt.xticks(rotation = 0)
plt.legend(title = 'Dejo la empresa?')
plt.tight_layout()
plt.show()
```
